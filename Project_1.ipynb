```python
# importing pandas library
import pandas as pd
```

```python
# Read the CSV File Dataset
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
# Display first 10 Rows of the Data Set
df.head(10)
```

```python
#Calculate Null Values in Each Column
df.isnull().sum()
```

```python
# Total Non Null values in Each Column
df.count()
```

```python
# Data Types of Each Column
print(df.dtypes)
```

```python
# Cleaning out 'Mean Score' column with value 's'

count_s_values = df['Mean Score'].value_counts(dropna=False)['s']
print("Number of rows with 's' in the 'mean score' column:", count_s_values)
```

```python
# Cleaning out 'count_number_scoring_below_65' with value 's'

count_number_scoring_below_65 = df['Number Scoring Below 65'].value_counts(dropna=False)['s']
print("Number of rows with 's' in the 'Number Scoring Below 65' column:",count_number_scoring_below_65 )
```

```python
# Cleaning out 'count_percent_scoring_below_65' with value 's'

count_percent_scoring_below_65 = df['Percent Scoring Below 65'].value_counts(dropna=False)['s']
print("Number of rows with 's' in the 'Percent Scoring Below 65' column:",count_percent_scoring_below_65)
```

```python
# Cleaning out the 's' Value from the below list of columns

columns_with_s_values = ['Mean Score', 'Number Scoring Below 65', 'Percent Scoring Below 65',
                         'Number Scoring 65 or Above', 'Percent Scoring 65 or Above',
                         'Number Scoring 80 or Above', 'Percent Scoring 80 or Above']

# Replace 's' values with NaN (missing values)
df[columns_with_s_values] = df[columns_with_s_values].replace('s', float('nan'))
```

```python
# Dropping the Rows with NaN values

df.dropna(inplace=True)
```

```python
# Checking count of Non Null Values after dropping

df.count()
```

```python
df.head(20)
```

```python
df.tail(30)
```

```python
# Cleaning out the 2 columns with 'na' values
columns_with_na_values = ['Number Scoring CR','Percent Scoring CR']

# Replace 'na' values with NaN (missing values)
df[columns_with_na_values] = df[columns_with_na_values].replace('na', float('nan'))
```

```python
# Dropping the rows with 'na' value

df.dropna(inplace=True)
```

```python
df.count()
```

```python
df.head()
```

```python
# Converting the Non-Numeric Columns to Numeric for finding Mean, Median, Max etc.

columns_to_convert = ['Mean Score', 'Number Scoring Below 65', 'Percent Scoring Below 65',
                      'Number Scoring 65 or Above', 'Percent Scoring 65 or Above',
                      'Number Scoring 80 or Above', 'Percent Scoring 80 or Above',
                      'Number Scoring CR', 'Percent Scoring CR']

# Convert the columns to numeric data type
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric)

# Verify the data types after conversion
print(df.dtypes)
```

```python
df.head()
```

```python
# Group by 'School DBN' and 'School Name', and calculate the average 'Mean Score' for each school
average_scores = df.groupby(['School DBN', 'School Name'])['Mean Score'].mean()

# Sort schools by their average scores in descending order
sorted_scores = average_scores.sort_values(ascending=False)
print(sorted_scores)
```

```python
# Picking Up 'School DBN': '20K187' as initial comparison Point
selected_dbn = '20K187'

selected_school = df[df['School DBN'] == selected_dbn]

# Print the information of the selected school
print("Selected School:")
print(selected_school)

# Comparing the mean score of the selected school with the overall mean score for all schools
overall_mean_score = df['Mean Score'].mean()
selected_school_mean_score = selected_school['Mean Score'].values[0]

print("\nOverall Mean Score for All Schools:", overall_mean_score)
print("Mean Score for the Selected School:", selected_school_mean_score)

```

```python
selected_dbn = '20K187'

# Selecting data for the selected school and schools with the same School Level
selected_school = df[df['School DBN'] == selected_dbn]
schools_same_level = df[df['School Level'] == selected_school['School Level'].values[0]]

# Calculating metrics for the selected school and schools with the same School Level
selected_school_mean_score = selected_school['Mean Score'].values[0]
same_level_mean_score = schools_same_level['Mean Score'].mean()

```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.bar(['Selected School', 'Same School Level Average'], [selected_school_mean_score, same_level_mean_score])
plt.ylabel('Mean Score')
plt.title(f'Comparison: Selected School vs. Schools in {selected_school["School Level"].values[0]} Level')
plt.show()
```

```python
# Calculating the mean, median, max and min score for the selected school and schools with the same School Level

selected_school_mean_score = selected_school['Mean Score'].values[0]
selected_school_min_score = selected_school['Mean Score'].min()
selected_school_max_score = selected_school['Mean Score'].max()
selected_school_median_score = selected_school['Mean Score'].median()

same_level_mean_score = schools_same_level['Mean Score'].mean()
same_level_min_score = schools_same_level['Mean Score'].min()
same_level_max_score = schools_same_level['Mean Score'].max()
same_level_median_score = schools_same_level['Mean Score'].median()


print("\nSelected School Statistics:")
print("Min Score:", selected_school_min_score)
print("Max Score:", selected_school_max_score)
print("Median Score:", selected_school_median_score)

print("\nSchools with Same Level Statistics:")
print("Min Score:", same_level_min_score)
print("Max Score:", same_level_max_score)
print("Median Score:", same_level_median_score)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(['Selected School', 'Same School Level Average'], [selected_school_mean_score, same_level_mean_score], label='Mean Score')
plt.errorbar(['Selected School'], selected_school_mean_score, yerr=[[selected_school_mean_score - selected_school_min_score], [selected_school_max_score - selected_school_mean_score]], fmt='o', capsize=10, color='red', label='Min/Max Range')
plt.errorbar(['Same School Level Average'], same_level_mean_score, yerr=[[same_level_mean_score - same_level_min_score], [same_level_max_score - same_level_mean_score]], fmt='o', capsize=10, color='red')
plt.ylabel('Mean Score')
plt.title(f'Comparison: Selected School vs. Schools in {selected_school["School Level"].values[0]} Level')
plt.legend()
plt.show()
```

```python
# Conclusion
# In this analysis, we compared the mean score of a selected school with other schools at the same School Level.
#The selected school performed higher than the average mean score of schools in its level.
```
